## Introduction

**What is Exploratory Data Analysis ?**

- Exploratory Data Analysis (EDA) is the process of analyzing and visualizing datasets to uncover patterns, trends, and relationships within the data before applying statistical modeling. Its primary goals include understanding the data structure, identifying anomalies and missing values, and generating hypotheses for further analysis. Common techniques used in EDA include descriptive statistics, data visualization (such as histograms and scatter plots), and correlation analysis. EDA is essential for gaining insights and guiding data-driven decisions in any analysis project.

**How to perform Exploratory Data Analysis ?**

- To perform Exploratory Data Analysis (EDA), start by importing necessary libraries (like Pandas and Matplotlib) and loading your dataset. Understand the data structure using methods like .head(), .info(), and .describe(). Check for missing values with .isnull().sum(), and clean the data by addressing duplicates or inconsistencies. Use visualizations, such as histograms, scatter plots, and box plots, to explore distributions and relationships between variables. Finally, summarize your findings to generate insights that can guide further analysis.

## 12. Outlier Detection
- For numerical columns, plot distributions using histograms, box plots, or sns.pairplot() to detect potential outliers visually. Calculate the Interquartile Range (IQR) or Z-scores for statistical outlier detection.

```python

from matplotlib import pyplot as plt 


x = df_last["Engine HP"]
y = df_last["MSRP"]

fig = plt.figure() 

ax = fig.add_subplot(111)

ax.plot(x, y, color = "green", linewidth = 5, linestyle ="-", label = "Price")

plt.legend()
plt.grid(True)
plt.title("My first diagram")
plt.xlabel("Engine HP")
plt.ylabel("Price")


plt.show()
```

---

## 13. Outlier Handling 
- Address outliers using methods like capping (clipping to upper/lower bounds), removing, or binning values. Consider keeping outliers if they represent meaningful data. Otherwise dropping the missing or null values.

```python
Q1 = df_last["MSRP"].quantile(0.25) 
Q3 = df_last["MSRP"].quantile(0.75)  
IQR = Q3 - Q1  

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Address outliers
outliers_ = df_last[(df_last["MSRP"] < lower_bound) | (df_last["MSRP"] > upper_bound)]
print("Outliers:\n", outliers_)
```

## 14. Exploratory Visualization:
- Plot histograms for distribution, bar plots for categorical variables, and pair plots to observe relationships between features. Use sns.heatmap() for visual correlation analysis between numerical columns.

```python
import seaborn as sns

df_number = df_last.select_dtypes(include="number")
plt.figure(figsize=(10, 8))
correlation_matrix = df_number.corr()  
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Korelasyon Is Haritas")
plt.show()
```
